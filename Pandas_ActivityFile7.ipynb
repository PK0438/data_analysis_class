### **More on DataFrames**

```python
import numpy as np
import pandas as pd
```

```python
random_data = np.random.randint(1, 101, (3, 5))
print(random_data)
```

```python
row_labels = ["Morning", "Afternoon", "Evening"]
column_labels = ["Monday", "Tuesday", "Wednesday", "Thursday", "Firday"]

df = pd.DataFrame(data=random_data, index=row_labels, columns=column_labels)
df
```

```python
df.transpose()          # transpose changes rows to columns and columns to rows
df.T                    # both same
```

```python
# nba dataset

nba = pd.read_csv("data/nba.csv", parse_dates=["Birthday"], index_col="Name")
nba.head(5)
```

```python
nba.dtypes.value_counts()
```

```python
nba.min()
```

```python
nba.max()
```

```python
nba.nlargest(n=3, columns="Salary")             # get top 3 maximum from single column Salary
```

```python
nba.nsmallest(n=3, columns="Salary")
```

```python
nba.sum(numeric_only=True)              # mean(), median(), mode(), std()
```

```python
nba.mean(numeric_only=True)
```

```python
nba.median(numeric_only=True)
```

```python
nba.mode(numeric_only=True)
```

```python
nba.std(numeric_only=True)
```

#### **Sorting by Index**
- axis = o is for sorting rows
- axis = 1 or "columns" is for sorting columns

```python
nba.sort_index(axis=1, ascending=False)          # axis=1 or axis="columns" for sorting columns, 

# by default ascending=True, to get descending order, have to explisitly turns ascending=False
```

```python
nba.sort_index(axis=0, ascending=False)          # sorting rows
```

#### **Select the rows and columns from a DataFrame**

```python
nba = nba.reset_index()                 # resets the index back as original
nba
```

```python
nba = nba.set_index("Name")
nba
```

```python
nba.dtypes
```

```python
nba.select_dtypes(include="object")         # includes only object data types
```

```python
nba.select_dtypes(exclude="int")            # excludes int data types
```

```python
nba.select_dtypes(exclude=["int", "object"])                # excludes int and object datatypes
```

```python
nba
```

```python
nba.loc[["Derrick Rose", "Harry Giles"], ["Team", "Salary"]]        # get the required rows with index and column names 

# loc: location, iloc: index location
```

```python
nba.loc["Christian Wood":"Marial Shayok",["Team", "Salary"] ]           # range from index Christian Wood to Marial Shayok, with the required columns
```

```python
nba.iloc[1, 2]          

#iloc = index location
# [row index, column index]
# row 1 = Christian Wood	
# column 2 = Birthday 
# which is Christian Wood's Birthday -> 1995-09-27
```

```python
nba = nba.reset_index()
nba
```
