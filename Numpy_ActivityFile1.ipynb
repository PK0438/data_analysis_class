## **Special Arrays**

```python
# Zeros
arr0 = np.zeros((3, 5))     # an array with 3x5 shape and all values as 0

print(arr0)
```

```python
arr0.dtype
```

```python
arr0 = np.zeros((3, 5), dtype='int')           # changing the data type of the elements from float to int

print(arr0)
```

```python
arr0.dtype
```

```python
# Ones

arr1s = np.ones((2, 8), dtype='float')          # An arrya with 2x8 shape and all the values are 1.0

print(arr1s)
```

```python
# full

arrfull = np.full((4, 4), 7)                    # An array with 4x4 shape and all values as 7

print(arrfull)
```

```python
# Identity matrix - diagonal elements are 1s and remaining all are 0

arrId = np.eye(4, dtype='int')
print(arrId)
```

```python
# array sequence

arrSeq = np.arange(5)

print(arrSeq)
```

```python
arrSeq2 = np.arange(2, 10, 2)
print(arrSeq2)
```

```python
# Equally spaced values
# np.set_printoptions(precision=3)  # Set display precision to 3 decimal places
equally_spaced_values = np.linspace(1, 3, 20)

print(equally_spaced_values)
```

```python
eq_sp_vl = np.linspace(5, 10, 10)

print(eq_sp_vl)
```

```python
# random distribution

arr_rand = np.random.rand(3,4)  # 3x4 array with random values between 0(inclusive) and 1(exclusive) [0, 1)

print(arr_rand)
```

```python
#

arr2_rand = np.random.rand(2, 3)

print(arr2_rand)
```

```python
#

arr_randint = np.random.randint(5, 10, size=(3, 7))
print(arr_randint)
```

```python
arr_randint2 = np.random.randint(5, 10, size=(3, 7))
print(arr_randint2)
```

```python
arr_randint3 = np.random.randint(5, 10, size=(4, 20))   # 3x7 array with random integers between 5 and 10 [5, 10)
print(arr_randint3)
```
