### **GroupBy Object**

```python
import numpy as np
import pandas as pd
```

```python
fortune = pd.read_csv("data/fortune1000.csv")
fortune
```

```python
# there are several ways to group this dataset, by sector, by industry, by profits etc

sectors = fortune.groupby("Sector")
sectors
```

```python
sectors.groups
```

```python
type(sectors.groups)
```

```python
sectors.first()

# first occurance of all the sectors 
```

```python
sectors.last()
```

```python
sectors.nth(5)

# nth row for each group. here 5th row in each group
```

```python
sectors.head()
```

```python
sectors.tail()
```

```python
sectors.sum()
```

```python
sectors.get_group("Apparel")
```

```python
sectors.get_group("Apparel").sum()
```

```python
sectors["Revenues"].sum()
```

```python
sectors["Profits"].sum()
```

```python
# group by 2 or more 

sectors_industries = fortune.groupby(["Sector", "Industry"])
sectors_industries.groups
```

```python
sectors_industries.first()
```

```python
# for sectors group by

agrregations = {
    "Profits": "sum",
    "Revenues": "mean",
    "Employees": "count"
}

sectors.agg(agrregations)
```
