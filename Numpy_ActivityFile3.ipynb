## **Array Operations and Broadcasting**

```python
import numpy as np

a1 = np.array([1, 2, 3])
a2 = np.array([4, 5, 6])
```

```python
print(a1 + a2)
print(a1 - a2)
print(a1 * a2)
print(a1 / a2)
```

```python
# Broadcasting - adding scalar to array

a1 + 10
```

```python

arr2 = np.array([1, 4, 9, 16])

print(np.sqrt(arr2))        # Square root of each element
print(np.log(arr2))        # Natural logarithm of each element
print(np.exp(arr2))        # Exponential of each element
print(np.sin(arr2))        # Sine of each element (in radians)
print(np.cos(arr2))        # Cosine of each element (in radians)
```
