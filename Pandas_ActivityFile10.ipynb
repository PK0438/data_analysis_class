## **Multi Index DataFrame**

```python
import numpy as np
import pandas as pd
```

```python
columns = ["Morning", "Evening"]
rows = ["Mon", "Tue", "Wed"]

data = [
    (1, 2),
    (3, 4),
    (5, 6)
]

pd.DataFrame(data=data, index=rows, columns=columns)
```

```python
columns_index = pd.MultiIndex.from_tuples([
    ("State", "Schools"),
    ("State", "Police"),
    ("Culture", "Museum"),
    ("Culture", "Restuarants")
])
rows = ["Mon", "Tue", "Wed"]

data = [
    (1, 2, 3, 4),
    (3, 4, 5, 6),
    (5, 6, 7, 8)
]

pd.DataFrame(data=data, index=rows, columns=columns_index)
```
