### **1. Element-wise Arithmetic**

**Between arrays:**

```python
import numpy as np

a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

print(a + b)
print(a - b)
print(a * b)
print(a / b)
print(a ** 2)
```

**With scalars:**

```python
print(a + 10)
print(a * 2)
```

### **2. Universal Functions (ufuncs)**

NumPy has **vectorized math functions** that apply element-wise:

```python
arr = np.array([1, 4, 9, 16])
np.sqrt(arr)
np.exp(arr)
np.log(arr)
np.sin(arr)
np.cos(arr)
```

### **3. Aggregation Functions**

```python
arr = np.array([[1, 2, 3], [4, 5, 6]])

print(arr.sum())
print(arr.sum(axis=0))
print(arr.sum(axis=1))

print(arr.mean())
print(arr.min())
print(arr.max())
print(arr.std())
```

### **4. Broadcasting (Different Shapes)**

Broadcasting lets Numpy do arithematic between arrays of different shapes. 

```python
arr = np.array([[1, 2, 3], 
                [4, 5, 6]])

add_vector = np.array([10, 20, 30])

print(arr + add_vector)
```

Even scalar + array works because scalars are treated as having shape `(1,)`

### **5. Comparison Operators**

```python
arr = np.array([1, 2, 3, 4])
print(arr > 2)
print(arr == 3)
```

**Any & All:**

```python
print(np.any(arr > 3))
print(np.all(arr > 0))
```

### **6. Clipping Values**

```python
arr. np.array([1, 5, 8, 10])
np.clip(arr, 3, 8)
```
