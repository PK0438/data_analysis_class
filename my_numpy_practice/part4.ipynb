### **1. Conditional Selection (`where`)**

Syntax: 
    np.where(condition, value_if_true, value_if_false)

```python
import numpy as np

arr = np.array([10, 20, 30, 40, 50])
result = np.where(arr > 25, 1, 0)
print(result)
```

### **2. Extracting Non-zero Elements**

```python
arr = np.array([10, 0, 42, 0, 3, 0, 0, 56])
indices = np.nonzero(arr)                               # non zero indices
print(indices)

print(arr[indices])                         # print non zero elements
```

### **3. Sorting Arrays**

```python
arr = np.array([1, 5, 3, 5, 4, 8, 6])
print(np.sort(arr))
```

```python
arr2d = np.array([[3, 1, 2], [10, 5, 9]])
print(np.sort(arr2d))

print(np.sort(arr2d, axis=1))
```

**Argsort (indices of sorted order):**

```python
print(arr)
print(np.sort(arr))
print(np.argsort(arr))


print(arr2d)
print(np.sort(arr2d))
print(np.argsort(arr2d))


```
