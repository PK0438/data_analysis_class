## **Working with Text Data Sets**

```python
import numpy as np
import pandas as pd
```

```python
inspections = pd.read_csv("data/chicago_food_inspections.csv")
```

```python
inspections.info()
```

```python
inspections
```

```python
inspections["Name"]
```

```python
inspections["Risk"]
```

```python
#writing for loop to strip the uneven spaces leading and trailing in the data set from all the columns. 

for column in inspections.columns:
    inspections[column] = inspections[column].str.strip()
```

```python
inspections["Name"]
```

```python
inspections["Risk"].unique()
```

```python
# dropping the nan rows

inspections = inspections.dropna(subset=["Risk"])
```

```python
inspections["Risk"].unique()
```

```python
inspections.shape       # the rows has been decreased   
```

```python
# replacing the Risk == all to risk 4(extream)

inspections = inspections.replace(
    to_replace="All",
    value="Risk 4 (Extream)"
)

inspections["Risk"].unique()
```

```python
# updating the risk column with only the severity by slicing the other characters

# inspections["Risk"] = inspections["Risk"].str[8:-1]
inspections["Risk"] = inspections["Risk"].str.slice(8, -1)

inspections.head()
```

```python
# display all rows from the Name column that contain "Pizza" (case-insensitive)
inspections[inspections['Name'].str.lower().str.contains('pizza')]
```

```python
# restuarent names that starts with 'tacos'

inspections[inspections["Name"].str.lower().str.startswith("tacos")]
```

```python
# another dataset

customers = pd.read_csv("data/customers.csv")

customers.head()
```

```python
# split the name column into two, first name and last name. 

customers["Name"].str.split(" ")
```

```python
# observe index 3 name. it has 3 parts. need only 2 names.
# pattern is " " (space), n = 1 is the first occurance of the pattern (space)

customers["Name"].str.split(pat=" ", n=1)

# now observe the name at 3rd index. 
```

```python
# splitting into two columns

customers["Name"].str.split(pat=" ", n=1, expand=True).get(0)
```

```python

customers["Name"].str.split(pat=" ", n=1, expand=True)[0]
```

```python
customers[["First Name", "Last Name"]] = customers["Name"].str.split(pat=" ", n=1, expand=True)

customers
```

```python
customers = customers.drop(
    labels="Name",
    axis=1
)

customers
```

```python

# customers["Address"].str.split(pat=",", expand=True)

customers[["Street", "City", "State", "Zip"]] = customers["Address"].str.split(pat=",", expand=True)
customers
```

```python
customers = customers.drop(
    labels="Address",
    axis=1
)
```
