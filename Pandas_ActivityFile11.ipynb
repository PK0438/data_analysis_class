## **Pivot Table**

```python
import numpy as np
import pandas as pd
```

```python
sales = pd.read_csv("data/sales_by_employee.csv", parse_dates=["Date"])
sales
```

```python
sales.pivot_table(
    index="Name",
    columns="Date",
    values="Revenue",
    aggfunc="sum"
)
```

```python
sales.pivot_table(
    index="Name",
    values="Revenue",
    aggfunc="sum"
)
```

```python
sales.pivot_table(
    index=["Name", "Date"],
    values="Revenue",
    aggfunc="sum"
)                                                               # each sales person sales wrt date
```

```python
sales.pivot_table(
    index=["Date", "Name"],
    values="Revenue",
    aggfunc="sum"
)     
```

```python
sales.pivot_table(
    index="Name",
    values=["Revenue", "Expenses"],
    aggfunc="sum"
)     
```

```python
sales.pivot_table(
    index="Name",
    values=["Revenue", "Expenses"],
    aggfunc="mean"
)     
```

```python
sales.pivot_table(
    index="Name",
    values=["Revenue", "Expenses"],
    aggfunc="min"
)     
```

```python
sales.pivot_table(
    index="Name",
    values=["Revenue", "Expenses"],
    aggfunc="max"
)     
```

```python
sales.pivot_table(
    index="Name",
    columns="Customer",
    values=["Revenue", "Expenses"],
    aggfunc="sum"
)     
```

```python
sales.pivot_table(
    index="Name",
    columns="Customer",
    values="Revenue",
    aggfunc="min"
)     
```

```python
sales.pivot_table(
    index="Name",
    columns="Customer",
    values="Revenue",
    aggfunc=["min", "count"]
)     
```

```python
sales.pivot_table(
    index="Name",
    columns="Customer",
    values="Revenue",
    aggfunc="sum",
    fill_value=0,
    margins=True,
    margins_name="Total"
)     
```

```python
# create two simple pivot tables. one is A and another is B

p1 = sales.pivot_table(
    index="Name",
    columns="Customer",
    values="Revenue",
    aggfunc="sum"
)    

p2 = sales.pivot_table(
    index=["Name", "Date"],
    values="Revenue",
    aggfunc="sum"
)  

# stack(), unstack()
```

```python
p1.stack()
```

```python
p2.unstack()
```

### **Melting a Dataset**

- unpivot the pivoted data is called melting a dataset

```python
# dataset = video game sales

vgs = pd.read_csv("data/video_game_sales.csv")
vgs
```

```python
vgs.melt(id_vars="Name", value_vars="NA")
```

```python
vgs.melt(id_vars="Name", value_vars=["NA", "EU", "JP", "Other"])
```

```python
vgs.melt(id_vars="Name", value_vars=["NA", "EU", "JP", "Other"],
         var_name="Region", value_name="Sales")
```

### **Exploding Dataset**
